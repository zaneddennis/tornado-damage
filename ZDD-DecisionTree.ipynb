```python
import pandas as pd
```

```python
df = pd.read_csv("Cleaned-Tornadoes_v2.csv", index_col="EVENT_ID")
df
```

```python
df = df.drop(columns=["DAMAGE_PROPERTY"])
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
df.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
X_columns = ["TOR_LENGTH", "TOR_WIDTH", "AVG_LAT", "AVG_LON", "YEAR", "AVG_RANGE", "Pop_Density", "sin_time", "cos_time"]
y_column = "CASUALTIES"
```

```python
test_df = df.sample(frac=0.2)
X_test = test_df.loc[:, X_columns]
y_test = test_df[y_column]
```

```python
training_df = df[~(df.index.isin(test_df.index))]
X_train = training_df.loc[:, X_columns]
y_train = training_df[y_column]
```

```python
reg = DecisionTreeRegressor().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```
